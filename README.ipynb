# Mini Project 2 Protein Design

# Background

De novo protein design is a cutting-edge field that aims to create entirely new protein structures and functions from scratch, without direct use of naturally occurring protein templates. With advances in computational methods, machine learning, and synthetic biology, the field is experiencing rapid progress, enabling the design of proteins with potential applications in therapeutics, industrial processes, and even artificial life forms. In this class we are using modern techniques to attempt creating our own protein design. This protein ideally would bind to the extracellular EGFR domain on the PDB protein 6ARU.

# Methods

To design this protein the most recent version of the public RF diffusion colab was used. RFdiffusion works by using a diffusion model to design proteins with specified shapes or binding capabilities. Starting with a random initial protein structure, it iteratively applies noise to the structure, progressively moving it toward a target conformation. At each step, the model adds controlled "noise" to simulate a random walk away from the initial state, then uses deep learning to denoise it, guiding it closer to the target structure in a smooth, continuous manner. By gradually reducing the noise over multiple iterations, RFdiffusion transforms an initial random or noisy structure into a well-defined protein that closely resembles the desired shape or functional features. This process enables the generation of novel protein backbones and sequences that can adopt stable, functional conformations, even if they differ significantly from natural proteins. The notebook used here runs RF diffusion taking several inputs such as contigs, PDB, iterations, hotspots, chains, and more. The notebook then performs protein MPNN and Alpha Fold to validate the generated backbone.

This protein was generated using the PDB 6ARU. The contigs input was set to 375, the selected chain A, and the hot spots were A11,A12,A13,A15,A16,A17,A18. The contig was set to 375 as a longer amino acid chain can enhance binding in proteins by increasing the number of potential interactions with a target. This allows for greater binding affinity, specificity, and structural complexity. In PDB structure 6ARU, residues 11, 12, 13, 15, 16, 17, and 18 are part of the interaction interface between the EGFR and Cetuximab, an antibody fragment designed to bind EGFR. Residues in this binding region often exhibit hydrophobic or polar side chains that enhance intermolecular forces, stabilizing the Cetuximab-EGFR complex. Therefore, it would be advantageous to design a protein targeting these regions of similar hydrophobic and hydrophilic properties. As they represent regions where modifications can optimize binding without compromising the proteins native structure. Thus those regions were chosen as targeted hotspots.
